### Photocure Enthalpy (J/g)

```python
Photocure_Enthalpy_df = data[['A','B','C','D','E','F','Photocure Enthalpy (J/g)']]
Photocure_Enthalpy_df.head()
#Criando um novo DF
```

```python
Photocure_Enthalpy_df.rename(columns={'Photocure Enthalpy (J/g)': 'Enthalpy'}, inplace=True)
Photocure_Enthalpy_df.head()
#Renomeando a coluna da variável resposta
```

```python
Photocure_Enthalpy_df.isnull().sum()
#Verificando valores nulos
```

```python
Photocure_Enthalpy_df.dropna(inplace=True)
#Retirando os valores nulos
```

```python
Photocure_Enthalpy_df.isnull().sum()
#Confirmando a remoção
```

```python
#Definindo atributos (variáveis independentes) e target (variáveis dependentes)
x2 = Photocure_Enthalpy_df[['A','B','C','D','E','F']]
y2  = Photocure_Enthalpy_df['Enthalpy']
```

```python
#Regressão

x = sm.add_constant(x2)
modelo_Enthalpy = sm.OLS(y2,x).fit()
modelo_Enthalpy.summary()
```
