##### Criei um dicion√°rio para modificar os status 'Alone' e 'YOLO' para o status 'Single'.

```python
dict_marital_status = {
        'Alone': 'Single',
        'Absurd': 'Absurd',
        'YOLO': 'Single',
        'Single': 'Single',
        'Together': 'Together',
        'Married': 'Married',
        'Divorced': 'Divorced',
        'Widow': 'Widow'
}
```

```python
df['New_Marital_Status'] = df['Marital_Status']
df['Marital_Status'] = df['Marital_Status'].map(dict_marital_status)
```

```python
df['Marital_Status'].unique()
```

##### Criei um dataframe (df_mean) com as m√©dias da coluna 'Income', usando os crit√©rios 'Marital_Status','Education'.

```python
df_mean = df.groupby(['Marital_Status','Education'])['Income'].agg(['mean'])
```

```python
df_mean.reset_index()
```

##### Criei um datraframe (df_updated) para receber o preenchimento dos valores nulos da coluna 'Income'. Usei o m√©todo merge para unificar o df_mean com o df_updated.

```python
df_updated = df.merge(df_mean, how='left', on=['Marital_Status','Education'])
```

```python
df_updated['Income'] = df_updated['Income'].fillna(df_updated['mean'])
```

```python
df_updated['Income'].isnull().sum()
```

##### Dropei a coluna 'mean' do dataframe df_updated.

```python
df_updated = df_updated.drop('mean', axis='columns')
```

```python
import matplotlib.pyplot as plt
```

```python
df_teste = df_target.groupby('Income_Classes')[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum().reset_index().sort_values('Income_Classes', ascending=False)

df_teste['Total'] = df_teste.loc[:,['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(axis=1)

df_teste.sort_values(by='Total', ascending=False).plot.bar(x='Income_Classes', y='Total')
```

```python
df_teste = df_target.groupby('Income_Classes')[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum().reset_index().sort_values('Income_Classes', ascending=False)

df_teste = df_teste[['Income_Classes','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']]
df_teste.set_index('Income_Classes', inplace=True)
df_total = df_teste['NumWebPurchases'] + df_teste['NumStorePurchases'] + df_teste['NumCatalogPurchases']
df_rel = df_teste[df_teste.columns].div(df_total, 0) * 100
df_rel.reset_index(names='Income_Classes', inplace=True)

fig, ax = plt.subplots(figsize=(8, 5))

df_rel.plot(x='Income_Classes', kind='barh', stacked=True, ax=ax)


for i, (row_label, row) in enumerate(zip(df_rel['Income_Classes'], df_rel.values)):
    cum_sum = 0
    for j, (category_label, category_value) in enumerate(zip(df_rel.columns[1:], row[1:])):
        value = category_value
        ax.text(cum_sum + value / 2, i, f"{value:.1f}%", ha='center', va='center', color='white', fontsize=10, fontweight='bold')
        cum_sum += value

ax.set_title('Stacked Bar Graph')
ax.set_xlabel('Percentage')
ax.set_ylabel('Income Classes')
ax.legend(fontsize = 6)
```

```python
product = df_target.groupby('Income_Classes')[
    'MntWines', 
    'MntFruits', 
    'MntMeatProducts', 
    'MntFishProducts', 
    'MntSweetProducts',
    'MntGoldProds'
].sum().reset_index().sort_values('Income_Classes', ascending=False)

product2 = product.drop('Income_Classes', axis=1).T.rename(columns={4:'A',3:'B',2:'C',1:'D',0:'E'})
product2['F'] = 1

fig, axs = plt.subplots(2,3,figsize=(20,10))

label = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
ar_aux = np.array(['A', 'B', 'C', 'D', 'E', 'F']).reshape((2,3))


for i in range(2):
    for j in range(3):     
        if ar_aux[i,j] != 'F':
            axs[i, j].pie(x=ar_aux[i,j],data=product2,autopct='%1.0f%%')
            axs[i, j].set_title(f'Products Consumption per Class {ar_aux[i,j]}')     
        else:
            axs[i, j].axis('off')

axs[1, 1].legend(fontsize=15, labels=label,loc='upper right', bbox_to_anchor=(2.4, 0.9))


fig.show()
    

```
