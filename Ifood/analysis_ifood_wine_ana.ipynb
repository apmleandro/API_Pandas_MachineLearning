```python
r= retail.groupby(by=['Education','Marital_Status'])['Income'].mean()
r
```

```python
# use transform to fill in the missing values with group-wise mean
retail['Income'] = retail.groupby(['Education', 'Marital_Status'])['Income'].transform(lambda x: x.fillna(r.loc[(x.name[0], x.name[1])])) #usar dicionario
```

## Plot Analysis ##

```python
retail['age']=retail['Year_Birth'].apply(lambda x : 2023-x) #tentar por ano

labels=['Gen Z','Millennials','Gen X','Boomers'] #tirar gen z
bins=[0,26,42,58,77]

wine_age = pd.DataFrame()
wine_age['wine']=retail['MntWines']
wine_age['age']=retail['age']

wine_age['Relation']= pd.cut(wine_age['age'],
                        bins=[0,26,42,58,77],
                        labels=labels,
                        right=False)
group = wine_age.groupby(['Relation'])['wine'].size()
group.plot(kind='pie', y='points',
           autopct='%1.0f%%',
           colors = ['red', 'pink', 'steelblue'],
           title='Wine Consumption per Generation')
```

```python
label2 = ['Below Poverty Line','Low Middle Class', 'Middle Class','High Middle Class','Rich'] #base brasileira procurar salario brasileiro
bin2 = [0,20000,52000,156000,200000,500000]
wine_wage = pd.DataFrame()
wine_wage['wine'] = retail['MntWines'] 
wine_wage['wage'] = retail['Income'] 
wine_wage['Relation']= pd.cut(wine_wage['wage'],
                        bins=[0,20000,52000,156000,200000,500000],
                        labels=label2,
                        right=False)
group2 = wine_wage.groupby(['Relation'])['wine'].size()
group2.plot(kind='pie', y='points',
           autopct='%1.0f%%',
           colors = ['red', 'pink', 'steelblue'],
           title='Wine Consumption per Class')
```

```python
wine_educ= pd.DataFrame()
wine_educ['wine'] = retail['MntWines'] 
wine_educ['Education'] = retail['Education'] 
group3 = wine_educ.groupby(['Education'])['wine'].size()
group3.plot(kind='bar')
```

```python
wine_marit= pd.DataFrame()
wine_marit['wine'] = retail['MntWines'] 
wine_marit['Marital_Status'] = retail['Marital_Status'] 
group4 = wine_marit.groupby(['Marital_Status'])['wine'].size()
group4.plot(kind='bar')
```

### Plottimng all plots using subplot ###

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))
group.plot(kind='pie', y='points',
           autopct='%1.0f%%',
           colors = ['red', 'pink', 'steelblue'],
           title='Wine Consumption per Generation',
           ax=axes[0][0])
group2.plot(kind='pie', y='points',
           autopct='%1.0f%%',
           colors = ['red', 'pink', 'steelblue'],
           title='Wine Consumption per Class',
           ax=axes[1][1])
group3.plot(kind='bar',ax=axes[1][0],title='Wine Consumption per Education')
group4.plot(kind='bar',ax=axes[0][1],title='Wine Consumption per Marital Status')
fig.tight_layout(pad=3.0)

plt.show()
```
